Throughout the semester we will learn more about what plots to use and when.  But for now we need to have a fundemental understanding of plotting in python.

```python
#upload the penguins data
import seaborn as sns
penguins=sns.load_dataset("penguins")
penguins.head()
```

```python
#import matplotlib
import matplotlib.pyplot as plt
```

What is the distribution of Penguin Mass?

```python
#With matplotlib your overall object is a figure and with that object you have axes.
#if we just have one plot, we just have one axes.

# Step 1 Make you canvas with the figure command
# Step 2 Make your plot
# Step 3 Label your plot
# Step 4 Label your Canvas
# Step 5 Show your plot

plt.figure(figsize=(10,12))
plt.hist(penguins['body_mass_g'],edgecolor='black')
plt.title('Histogram of Penguin Body Mass')
plt.ylabel('Count of Penguins')
plt.xlabel('Body Mass in Grams')
plt.suptitle('The outside figure')
plt.show()
```

We want to make two graphs:
1. What is the relationship between flipper lenght and mass?

2. Does the average mass change by Species?

```python
average_mass=penguins.groupby('species')['body_mass_g'].mean().reset_index()
```

```python
average_mass
```

```python
# Step 1 Make you canvas and axeses with plt.subplots
# Step 2 Make your first plot
# Step 3 Label your plot

# Step 3 Make your first plot
# Step 4 Label your second plot
# Step 4 Label your Canvas
# Step 5 Show your plot
fig,(ax_scatter,ax_bar)=plt.subplots(1,2,figsize=(10,5))
ax_scatter.scatter(x=penguins['flipper_length_mm'],y=penguins['body_mass_g'])
ax_scatter.set_title('Penguins flip lenght vs mass')
ax_scatter.set_xlabel('Flipper Length in mm')
ax_scatter.set_ylabel('Body Mass in Grams')

ax_bar.bar(x=average_mass['species'],height=average_mass['body_mass_g'])
ax_bar.set_title('Average Mass by Species')
ax_bar.set_xlabel('Species')
plt.suptitle('More Penguins is better')
plt.show()

```

```python
df = pd.read_csv('/content/CPS_for_plotting.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'EARNWEEK2' and 'State_Name' columns are clean
earning_data_clean = df[['EARNWEEK2', 'State_Name']].dropna()

# Calculate average weekly earnings by state
average_earnings_by_state = earning_data_clean.groupby('State_Name')['EARNWEEK2'].mean().sort_values(ascending=False).reset_index()

# Create the bar plot with increased height for better readability
plt.figure(figsize=(16, 12)) # Increased height from 9 to 12
sns.barplot(x='EARNWEEK2', y='State_Name', data=average_earnings_by_state, palette='viridis', hue='State_Name', legend=False)

# Add labels and title
plt.title('Average Weekly Earnings by State', fontsize=20)
plt.xlabel('Average Weekly Earnings ($)', fontsize=15)
plt.ylabel('State', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Improve layout and display
plt.tight_layout()
plt.show()
```
